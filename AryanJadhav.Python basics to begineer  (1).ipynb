# You're getting this error:
RuntimeError: asyncio.run() cannot be called from a runnig event loop
This happens only in Jupyter Notebook/Google Colab,not in normal Python.

# üß† Why the Error Happens?

Jupyter Notebook is already running an async loop in the background.

So when you write:

Python says:


"Hey! I'm already running an event loop.You can't start another one."

So it gives RuntimeError.

````python
# 1 import math
import math
````

````python
# 2. Trigonometry functions
print(math.sin(0))
print(math.cos(0))
print(math.tan(0))

print(math.asin(1))
print(math.acos(1))
print(math.atan(1))
print(math.atan2(1,1))
````

````python
# 3 . Hyperbolic functions
print(math.sinh(1))
print(math.cosh(1))
print(math.tanh(1))

print(math.asinh(1))
print(math.acosh(2))
print(math.atanh(0.5))
````

````python
# 4 Rounding Function
print(math.ceil(5.2))
print(math.floor(5.8))
print(math.trunc(5.9))
````

````python

# 5 Logarithms
print(math.log(10))       # Natural logarithm (base e)
print(math.log10(10))     # Base 10 logarithm
print(math.log2(8))       # Base 2 logarithm
print(math.log1p(1))      # log(1 + x)
````

````python
# 6 Powers and Roots
print(math.sqrt(25))
print(math.pow(2, 3))
print(math.exp(1))
print(math.expm1(1))
print(math.exp2(3))
````

````python
# 7 Factorial, GCD , Permutation , Combination
print(math.factorial(5))
print(math.gcd(12,8))
print(math.perm(5,2))
print(math.comb(5,2))
````

````python
# 8 other useful Functions
print(math.fabs(-5))
print(math.fmod(7, 3))
print(math.modf(5.75))
print(math.copysign(5, -1))
print(math.dist([0,0],[3,4]))
print(math.hypot(3,4))
print(math.fsum([1,2,3]))
print(math.prod([2,3,4]))
print(math.remainder(10,3))
````

````python
# 9 Constants
print(math.pi)
print(math.e)
print(math.tau)
print(math.inf)
print(math.nan)
````

````python
# 10 Checking Numbers
print(math.isfinite(5))
print(math.isinf(math.inf))
print(math.isnan(math.nan))
print(math.isclose(1.00001, 1.0))
````
